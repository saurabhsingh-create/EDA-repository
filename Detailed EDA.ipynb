```python
#since there is a lot of missing values in age column we can impute using average
#instead of imputing age directly we can use passender wise avg age
#we can check this using a box plot
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
#get the averge age using filters
round(train[train['Pclass'] ==1]['Age'].mean()), round(train[train['Pclass'] ==2]['Age'].mean()), round(train[train['Pclass'] ==3]['Age'].mean())
```

```python
#function to apply on columns
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return round(train[train['Pclass'] ==1]['Age'].mean())

        elif Pclass == 2:
            return round(train[train['Pclass'] ==2]['Age'].mean())

        else:
            return round(train[train['Pclass'] ==3]['Age'].mean())

    else:
        return Age
```

```python
#apply the function on the columns
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
#check for nulls
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#train.isnull().sum()
```

```python
#drop columns with most nulls data and null rows in embarked
train.drop('Cabin', axis =1, inplace = True)
train.dropna(inplace=True)
train.head()
```

## Build a logistic regression model

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived',axis=1), 
                                                    train['Survived'], 
                                                    test_size=0.30, 
                                                    random_state=101)
```

```python
#training and predicting
from sklearn.linear_model import LogisticRegression
Logmodel = LogisticRegression()
Logmodel.fit(X_train, y_train)
```

```python
#to get the predictions on test data
predictions = Logmodel.predict(X_test)
```

```python
#to get the right evaluation parameters
from sklearn.metrics import confusion_matrix
accuracy = confusion_matrix(y_test, predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,predictions)
accuracy

predictions
```

```python
#we can print precision recall, F1 score using classfication report
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
#plotting decion trees in random forest
#Random forest is bagging technique internally it used number of decision trees runs on multiple subset of data, 
#picked by random sampling with replacement and takes average of all the predictions
#In ranodm forest n_estimatrs is the no of trees.
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10, random_state=0)
iris = load_iris()
X,y = iris.data, iris.target
clf = clf.fit(X,y)
```

```python
#to view all the estimators/trees
clf.estimators_
```

```python
#increase the plot size using figszie and plot using tree.plot_tree
plt.figure(figsize= (20,12))
tree.plot_tree(clf.estimators_[0], filled =True)
```

```python
#plotting all decison trees together
for i in range(len(clf.estimators_)):
    #print(i)
    #this uses only 1 canvas so it will replace all other and keep only one.
    plt.figure(figsize=(15,10))
    tree.plot_tree(clf.estimators_[i], filled =True)
```

```python
#image may not print all but all text can be printed for all decsion trees in aloop using export_text
for i in range(len(clf.estimators_)):
    print(tree.export_text(clf.estimators_[i]))
```
